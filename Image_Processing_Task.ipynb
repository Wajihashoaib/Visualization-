# Importing necessary libraries

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

# Reading the image

```python
img = cv2.imread("image.jpg")
```

# Applying Average Filtering

```python
blur = cv2.blur(img, (3,3))
```

# Displaying the result

```python
cv2_imshow(img)
cv2_imshow(blur)
```

# Applying Gaussian Filtering

```python
Gblur = cv2.GaussianBlur(img, (3,3), 0)    
```

# Displaying the result

```python
cv2_imshow(img)
cv2_imshow(Gblur)
```

# Applying Median Filtering

```python
Mblur = cv2.medianBlur(img, 3)
```

# Displaying the result

```python
cv2_imshow(img)
cv2_imshow(Mblur)
```

# Small and Large

```python
# Apply average filtering with a small mask (3x3) (-1 shows that output image will have the same no.of channels as input.)
average_small = cv2.filter2D(img, -1, np.ones((3, 3), np.float32)/9)
cv2_imshow(average_small)

# Apply average filtering with a large mask (9x9)
average_large = cv2.filter2D(img, -1, np.ones((9, 9), np.float32)/81)
cv2_imshow(average_large)

# Apply Gaussian filtering with a small mask (3x3)
gaussian_small = cv2.GaussianBlur(img, (3, 3), 0)
cv2_imshow(gaussian_small)

# Apply Gaussian filtering with a large mask (9x9)
gaussian_large = cv2.GaussianBlur(img, (9, 9), 0)
cv2_imshow(gaussian_large)

# Apply median filtering with a small mask (3x3)
median_small = cv2.medianBlur(img, 3)
cv2_imshow(median_small)

# Apply median filtering with a large mask (9x9)
median_large = cv2.medianBlur(img, 9)
cv2_imshow(median_large)
```

Here, we have applied each filter with both a small and a large mask. The small filter has a size of 3x3, and the large filter has a size of 9x9.
So, a small mask will smooth the image less than a large mask. This means that a small filter will preserve more of the details in the image, while a large filter will blur the image more.
