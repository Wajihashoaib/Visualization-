# Importing Libraries 

```python
import matplotlib.pyplot as plt
from matplotlib import style
```

# Lists 

```python
days = [2,3,4,5,6,7,8,9,12,20,28]
temp = [2.1, 2.4, 3.2 , 4.5 , 6.5,7.8,8.9,9.9,12.78,18.9,34.7]
```

# Graph

```python
plt.plot(days, temp , marker= 'o' , color = 'green', markersize = 13 )
style.use('ggplot')
plt.title('Temprature Graph' ,  fontsize = 18 , color = 'r')
plt.xlabel('Days' , fontsize = 16 , color = 'Blue' )
plt.ylabel('Temprature', fontsize = 16 , color = 'Blue' )
plt.legend('temp-line' , loc = 4)
plt.grid()
plt.show()
```

# Multiple Line Graphs

```python
days = [2,3,4,5,6,7,8,9,12,20,28]
temp_karachi = [2.1, 2.4, 3.2 , 4.5 , 6.5,7.8,8.9,9.9,12.78,18.9,34.7]
temp_Islamabad = [3.1, 7.4, 17.2 , 14.5 , 66.5,77.8,88.9,19.9,14.78,12.9,55.7]
```

```python
plt.plot(days, temp_karachi , marker= 'o' , color = 'green', markersize = 13 )
plt.plot(days, temp_Islamabad , marker= '^' , color = 'blue', markersize = 13 )
style.use('ggplot')
plt.title('Temprature Graph' ,  fontsize = 16 , color = 'r')
plt.xlabel('Days' , fontsize = 13 )
plt.ylabel('Temprature', fontsize = 14)
plt.legend('temp-line' , loc = 4)
plt.show()
```
