# Importing the necessary libraries

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

# Reading the image file

```python
image = cv2.imread('test.jpg')
```

# Converting the image to grayscale

```python
image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## Performing edge detection using Sobel operator to detect horizontal and vertical edges

```python
gradients_sobelx = cv2.Sobel(image, -1, 1, 0) # Detect edges in the x direction
gradients_sobely = cv2.Sobel(image, -1, 0, 1) # Detect edges in the y direction
gradients_sobelxy = cv2.addWeighted(gradients_sobelx, 0.5, gradients_sobely, 0.5, 0) # Combining both x and y edges
```

## Performing edge detection using Laplacian operator to detect edges in all directions

```python
gradients_laplacian = cv2.Laplacian(image, -1)
```

## Performing edge detection using Canny operator with threshold values 80 and 150 to detect edges

```python
canny_output = cv2.Canny(image, 80, 150)
```
